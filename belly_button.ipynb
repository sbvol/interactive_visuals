```python
import datetime as dt
import numpy as np
import pandas as pd
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
from flask import Flask, jsonify
from sqlalchemy import create_engine, MetaData, inspect
```

```python
#################################################
# Database Setup
#################################################
engine = create_engine("sqlite:///belly_button_biodiversity.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# Inpecting tables to ensure the create_all command worked
inspector = inspect(engine)
```

```python
inspector.get_table_names()
```

```python
inspector.get_columns('otu')
```

```python
inspector.get_columns('samples_metadata')
```

```python
Sam_Meta = Base.classes.samples_metadata
```

```python
results = session.query(Sam_Meta.SAMPLEID).all()
```

```python
bb_names=[]
for r in results:
    bb_names.append("BB_" + str(r[0]))
```

```python
OTU = Base.classes.otu
```

```python
results = session.query(OTU.lowest_taxonomic_unit_found).all()
```

```python
otu_desc=[]
for r in results:
    otu_desc.append(str(r[0]))
```
